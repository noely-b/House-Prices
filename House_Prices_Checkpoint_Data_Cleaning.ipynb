# Missing Values

```python
#Identify missing values
df.isnull().sum()
```

```python
#View missing values
df[df.isnull().any(axis=1)]
```

There's missing data on some regions for the 10 yr period. Given that the project focuses on urban areas, I would eliminate most of the regions with "NaN" values on the 1/31/2015 column; with exceptions being Hilo HI, Winnemucca NV  and Morgan City LA due to them having data that rounds up to 10 years.

```python
#Eliminate rows 205, 251, 337, 387, 428, 477, 509, 513, 551, 593, 596, 629, 639, 641, 690, 697, 708, 722, 746, 747, 761, 777, 795, 804, 810, 816, 821, 847, 851, 854, 856, 860, 863, 871, 872, 886, 888
# Create a list of rows to drop
rows_to_drop = [205, 251, 337, 387, 428, 477, 509, 513, 551, 593, 596, 629, 639, 641, 690, 697, 708, 722, 746, 747, 761, 777, 795, 804, 810, 816, 821, 847, 851, 854, 856, 860, 863, 871, 872, 886, 888]

# Create a boolean mask to identify rows to keep
rows_to_keep = ~df.index.isin(rows_to_drop)

# Filter the DataFrame to keep only the selected rows
df = df[rows_to_keep]
```

```python
#View missing values
df[df.isnull().any(axis=1)]
```

The remaining missing values can now be filled through median

```python
#Replace missing values with median
# Select numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

# Replace missing values with median for numeric columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
#View missing values
df[df.isnull().any(axis=1)]
```

```python
#Save dataset
df.to_csv('cleaned_HousePricesdataset.csv', index=False)
```

# National House Prices Trend

It is helpful to know the general house pricing trend. Therefore, this section will be attempting to answer: What's the national housing price per year?

```python
#Create a graph that represents national average housing price per year
# Assuming your DataFrame is named 'df' and the date columns start with '1/31/2015'
date_columns = df.columns[df.columns.get_loc('1/31/2015'):]

# Calculate the average price for each date (year)
average_prices = df[date_columns].mean()

# Create a new DataFrame for the average prices
average_prices_df = pd.DataFrame({'Date': average_prices.index, 'AveragePrice': average_prices.values})

# Create the line graph
fig = px.line(average_prices_df, x='Date', y='AveragePrice', title='National Average Housing Price per Year')

# Customize the graph (optional)
fig.update_layout(xaxis_title='Year', yaxis_title='Average Housing Price')

# Show the graph
fig.show()
```

The line chart represents the national house prices average per year. It gives us an idea on how the nation is performing house-pricing wise, and allows us to see a trend on how housing price is increasing. We can see that there was a sharp increase in house prices around Oct 2020, and it kept (generally speaking) increasing since then. This data also helps in identifying whether a state/region is below/above national data; and for any further analysis which can benefit from a general data reference
